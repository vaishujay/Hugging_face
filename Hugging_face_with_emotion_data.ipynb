```python
from datasets import load_dataset

# Load the 'emotion' dataset
emotion = load_dataset('emotion')

# Print the dataset
print(emotion)
```

```python
emotion   #check splitiong
```

**Set Dataframe format to pandas dataframe**

```python
emotion.set_format(type='pandas')
```

```python
df=emotion['train'][:]
df
```

```python
df.head()
```

```python
classes=emotion['train'].features['label'].names # label name from data
classes
```

```python
df['label_name']=df['label'].apply(lambda x: classes[x])
df.head()
```

***Data Analysis***

```python
import matplotlib.pyplot as plt
```

```python
df['label_name'].value_counts()
```

```python
label_counts=df['label_name'].value_counts(ascending=True)
label_counts.plot.barh()
plt.title('fequency of each label')
plt.show()
```
